## FINALMENTE, UNIMOS Y PROBAMOS SOBRE CHUNKS SOBRE EL ARCHIVO DE PRUEBA:

```python
import pandas as pd
from datetime import datetime as dttm

import numpy as np
pd.set_option('display.max_columns', None)

import matplotlib.pyplot as plt
%matplotlib inline

archivo_origen='searches.csv.bz2'
archivo_muestra= 'sample.'+archivo_origen[0:len(archivo_origen)-4]

#archivo_origen

archivo=archivo_muestra
#bz2archivo=archivo+'.bz2'

#!bunzip2 -k $bz2archivo

daf=pd.read_csv(archivo, sep='^')

df=daf[['Date','Time','Country','Destination']]

df['Datos_nuevos']=df['Date'].apply(lambda x: dttm.strptime(x,'%Y-%m-%d'))

df['Año']=df['Datos_nuevos'].apply(lambda x: x.year)
df['Mes']=df['Datos_nuevos'].apply(lambda x: x.month)
df['Dia']=df['Datos_nuevos'].apply(lambda x: x.day)


bool_array=np.logical_or(
                np.logical_or(df['Destination']=='MAD',  
                              df['Destination']=='BCN'),
                df['Destination']=='AGP')

resultados=df[bool_array].groupby(['Año','Mes','Destination'])['Dia'].count().reset_index()
resultados.rename(columns={'Dia':'Busquedas'}, inplace=True)
resultados



a=plt.plot(resultados['Mes'][resultados['Destination']=='MAD'],resultados['Busquedas'][resultados['Destination']=='MAD'])
b=plt.plot(resultados['Mes'][resultados['Destination']=='AGP'],resultados['Busquedas'][resultados['Destination']=='AGP'])
c=plt.plot(resultados['Mes'][resultados['Destination']=='BCN'],resultados['Busquedas'][resultados['Destination']=='BCN'])

plt.legend(['Madrid','Málaga','Barcelona'])
```

## INTRODUCIMOS LOS CHUNKS:

```python
import pandas as pd

from datetime import datetime as dttm

import numpy as np
pd.set_option('display.max_columns', None)

import matplotlib.pyplot as plt
%matplotlib inline

archivo_origen='searches.csv.bz2'

archivo_or=archivo_origen[0:len(archivo_origen)-4]
archivo_muestra= 'sample.'+archivo_origen[0:len(archivo_origen)-4]

#archivo_or
#archivo_muestra

archivo=archivo_or
bz2archivo=archivo+'.bz2'

print((archivo,bz2archivo))

#!bunzip2 -k $bz2archivo

daf=pd.read_csv(archivo, sep='^', chunksize=300000)

resultados=pd.DataFrame([])

for trozo in daf:
    
    trozo=trozo.dropna()

    df=trozo[['Date','Time','Country','Destination']].copy()
    
    df['Date']=df['Date'].apply(lambda x: str(x))
    df['Date']=df['Date'].apply(lambda x: x[0:10])

    df['Datos_nuevos']=df['Date'].apply(lambda x: dttm.strptime(x,'%Y-%m-%d'))

    df['Año']=df['Datos_nuevos'].apply(lambda x: x.year)
    df['Mes']=df['Datos_nuevos'].apply(lambda x: x.month)
    df['Dia']=df['Datos_nuevos'].apply(lambda x: x.day)


    filtrado=df[(df['Destination']=='MAD')| (df['Destination']=='BCN')| (df['Destination']=='AGP')]

    traspaso=filtrado.groupby(['Año','Mes','Destination'])['Dia'].count().reset_index()
    resultados=resultados.append(traspaso)
    
res=resultados.groupby(['Año','Mes','Destination'], as_index=False)['Dia'].sum()
    
res.rename(columns={'Dia':'Busquedas'}, inplace=True)
```
